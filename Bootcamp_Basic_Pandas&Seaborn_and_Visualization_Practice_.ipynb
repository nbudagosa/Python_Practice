# Calculation

```python
df[:3]
```

```python
df["total_ortherrooms"] = df["total_rooms"] - df["total_bedrooms"] #สร้าง total_ortherrooms จาก total_rooms - total_bedrooms
```

```python
df[:3][["total_ortherrooms", "total_rooms", "total_bedrooms"]] # เรียกดู 3 แถวแรกจาก total_ortherrooms
```

```python
df["median_house_value_vat"] =  df["median_house_value"] + (df["median_house_value"] * 7/100)
#สร้าง median_house_value_vat จาก median_house_value + (median_house_value * 7/100)
```

```python
df[["median_house_value_vat", "median_house_value"]][:3] #เรียกดู 3 แถวแรกของ median_house_value_vat และ median_house_value
```

```python
df["median_house_value_vat"].describe() # เรียกดูค่าสถิติเบื้องต้น
```

```python
# สร้าง function 
def transform_price(_price):

    if _price <= 127758:
        return "economics"
    elif _price >= 283550:
        return "luxury"
    else:
        return "market"
```

```python
df["price"] = df["median_house_value_vat"].apply(transform_price)
# สร้าง prince จาก median_house_value_vat .apply(transform_price)
```

```python
df[["price", "median_house_value_vat"]].sample(5) # เรียกดู column price,median_house_value_vat แบบสุ่มให้ผลออกมา 5 แถว
```

```python
condition_price = df["price"].str.contains("lux")
```

```python
df[condition_price][:5] # สร้างตาราง condition_price 5 แถวแรก
```

# Group by Aggregate

```python
df["price"].value_counts() # นับ
```

```python
# เรียกดู Aggregate
df.groupby(["price"])[["median_income"]].agg(["min", "mean", "median", "max", "std"])
```

```python
df["median_income"].describe()
```

```python
def transform_income(_income):

    if _income <= 2.5:
        return "poor"
    elif _income >= 4.7:
        return "rich"
    else:
        return "middle"
# สร้าง function
```

```python
df["income"] = df["median_income"].apply(transform_income)
```

```python
df[["income", "median_income"]].sample(5) # เรียกดูแบบสุ่ม income ,median_income
```

```python
df_group_price_income = df.groupby(["price", "income"])[["median_income"]]\
    .agg(["count", "min", "mean", "median", "max", "std"]) 
# สร้าง df_group_price_income 
```

```python
df_group_price_income.style.background_gradient() #สร้างตาราง style.background_gradient
```

# Pivotable

```python
df_pivot_price_income = pd.pivot_table(df, 
               values="median_income", 
               index="price", 
               columns="income", 
               aggfunc='mean', 
               fill_value=None)
# สร้าง df_pivot_price_income และกำหนดค่าต่างๆ
```

```python
df_pivot_price_income # เรียกดูที่สร้างข้างบน
```

```python
df_pivot_price_income.style.background_gradient() # ใส่ style.background_gradient ในตาราง
```
